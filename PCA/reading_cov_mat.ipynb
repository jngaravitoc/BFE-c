```python
def reading_cov_matrix(nmax, lmax):
    """
    Function that reads the covariance matrix.
    
    Input:
    ------
    nmax: int(nmax)
    lmax: int(lmax)
    
    output:
    -------
    A 2d array with the covariance matrix.
    
    """
    cov_mat_data = np.loadtxt('data/cov_mat_sph_5x5.txt', skiprows=1)
    i=0
    k=0
    j=0
    covariance_matrix = np.zeros([int((nmax+1)*(lmax+1)*((lmax+2)/2.0)), int((nmax+1)*(lmax+1)*((lmax+2)/2.0))])
    for i in range(int((nmax+1)*(lmax+1)*((lmax+2)/2.0))):
        for j in range(int((nmax+1)*(lmax+1)*((lmax+2)/2.0))):
            covariance_matrix[i][j] = cov_mat_data[k]
            k+=1

    return covariance_matrix
```

```python
matrix = reading_cov_matrix(5, 5)
```

```python
figure(figsize=(7,5))
imshow(np.log10(np.abs(matrix)), cmap='RdYlGn')
colorbar()
title('$Covariance\ matrix\ for\ (n_{max}=5,\ l_{max}=5,\ m_{max=5})$')
xlabel(r'$(n_{max}+1)\cdot(l_{max}+1) \cdot \frac{(m_{max}+2)}{2.0}$', fontsize=20)
ylabel(r'$(n_{max}+1)\cdot(l_{max}+1)\cdot \frac{(m_{max}+2)}{2.0}$', fontsize=20)
```
